```python
import matplotlib.pyplot as plt
import numpy as np

# Quadratic solver function (unchanged from previous code)
def quadratic_solver(a, b, c):
    delta = b**2 - 4*a*c
    if delta >= 0:
        x1 = (-b + (delta**0.5)) / (2*a)
        x2 = (-b - (delta**0.5)) / (2*a)
        return x1, x2
    else:
        return None

# Function to plot quadratic equation and solutions
def plot_quadratic_equation(a, b, c, solutions):
    # Generate x values
    x = np.linspace(-10, 10, 100)

    # Calculate y values for the quadratic equation
    y = a*x**2 + b*x + c

    # Plot the quadratic equation
    plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')

    # Mark the solutions on the plot
    if solutions:
        plt.scatter(solutions, [0, 0], color='red', marker='x', label='Solutions')

    # Add labels and legend
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.title('Quadratic Equation and Solutions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Show the plot
    plt.show()

# Example coefficients
a = 1
b = -3
c = 2

# Call the quadratic_solver function
solutions = quadratic_solver(a, b, c)

# Print the solutions
print("Solutions:", solutions)

# Plot the quadratic equation and solutions
plot_quadratic_equation(a, b, c, solutions)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Quadratic solver function (unchanged from previous code)
def quadratic_solver(a, b, c):
    delta = b**2 - 4*a*c
    if delta >= 0:
        x1 = (-b + (delta**0.5)) / (2*a)
        x2 = (-b - (delta**0.5)) / (2*a)
        return x1, x2
    else:
        return None

# Function to plot quadratic equation and solutions
def plot_quadratic_equations(equations):
    # Generate x values
    x = np.linspace(-10, 10, 100)

    # Plot each quadratic equation
    for equation in equations:
        a, b, c, solutions = equation

        # Calculate y values for the quadratic equation
        y = a*x**2 + b*x + c

        # Plot the quadratic equation
        plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')

        # Mark the solutions on the plot
        if solutions:
            plt.scatter(solutions, [0, 0], color='red', marker='x')

    # Add labels and legend
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title('Quadratic Equations and Solutions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Show the plot
    plt.show()

# User input loop
equations = []
while True:
    try:
        # Get coefficients from the user
        a_input = input("Enter coefficient a (or type 'exit' to stop): ")

        # Check if the user wants to exit
        if a_input.lower() == 'exit':
            print("Exiting the program.")
            break

        # Convert input to float
        a = float(a_input)
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))

        # Call the quadratic_solver function
        solutions = quadratic_solver(a, b, c)

        # Print the solutions
        print("Solutions:", solutions)

        # Save equation and solutions for plotting
        equations.append((a, b, c, solutions))

    except ValueError:
        print("Invalid input. Please enter valid numeric coefficients.")

# Plot all quadratic equations and solutions together
plot_quadratic_equations(equations)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(a, b, c, x):
    """
    Calculate the quadratic solution for the given coefficients and x value.
    """
    y = a * x**2 + b * x + c
    return y

def plot_temperature_over_time(a, b, c, time_range):
    """
    Plot temperature over time using a quadratic function.
    """
    time = np.array(time_range)
    temperature = quadratic_solution(a, b, c, time)

    plt.plot(time, temperature, label=f'a={a}, b={b}, c={c}')

def main():
    print("Weather Modeling using Quadratic Solution")

    # Get the number of sets
    num_sets = int(input("Enter the number of sets: "))

    for i in range(num_sets):
        # Input coefficients for each set
        if i == 0:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            c = float(input("Enter the value of c: "))
        else:
            a = -5
            b = 65
            c = 7

        # Hard-coded time values
        time_range = np.linspace(0, 10, 100)  # Adjust the time range as needed

        # Plot temperature over time using the quadratic function for each set
        plot_temperature_over_time(a, b, c, time_range)

    # Show the legend and display the plot
    plt.legend()
    plt.title("Temperature Over Time for Multiple Sets")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
1
```
